### 1) 알고리즘 작성 

```

2. 전체 이용 내역을 돌며 반복하기
   2.1. 각 이용내역에서 매입상태가 '전표매입'이라면
       2.1.1. 해당 건의 이용월과 이용 금액 구하기
       2.1.2. 이용월에서 10을 빼 월별 지출액 리스트의 인덱스 구하기
       2.1.3. 2.1.2.에서 구한 인덱스의 값에 과정 2.1.1.에서 구한 이용금액 더하기
3. 월별 지출액 리스트로 막대그래프 그리기
```
